```python
deck = list(range(108))
end_index = 107
lfsr = 0b1100100  # Initial seed
steps = 0

def update_lfsr(lfsr):
    bit = ((lfsr >> 3) ^ (lfsr >> 0)) & 1
    lfsr = (lfsr >> 1) | (bit << 6)
    return lfsr

# Main loop
while True:
    lfsr = update_lfsr(lfsr)
    rand_index = lfsr
    steps += 1
    temp = 0
    while(rand_index > end_index):
        lfsr = update_lfsr(lfsr)
        rand_index = lfsr
        steps += 1
        temp += 1

    print(f"More {temp} steps")
    # Swap the cards at the end of the deck and the random index
    deck[end_index], deck[rand_index] = deck[rand_index], deck[end_index]
    print(deck[end_index])
    end_index -= 1
    
    if end_index == 0:
        break

print(f"Shuffling took {steps} steps")
```
